```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
browser = Browser('chrome')
```

### Step 1: Visit the Website

Use automated browsing to visit the [Mars Temperature Data Site](https://static.bc-edx.com/data/web/mars_facts/temperature.html). Inspect the page to identify which elements to scrape.

   > **Hint** To identify which elements to scrape, you might want to inspect the page by using Chrome DevTools to discover whether the table contains usable classes.


```python
# Visit the website
# https://static.bc-edx.com/data/web/mars_facts/temperature.html
url = "https://static.bc-edx.com/data/web/mars_facts/temperature.html"
browser.visit(url)
```

### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create a Beautiful Soup object
html = browser.html  
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Extract all rows of data
# 1. Identify the table (replace 'table' with the actual tag or a unique class/id)
table = soup.find('table')

# 2. Extract all rows (assuming the rows are within <tr> tags)
rows = table.find_all('tr')

# 3. Extract data from each row (replace 'td' with the appropriate tag if different)
data = []
for row in rows:
    cols = row.find_all('td')
    cols = [col.text.strip() for col in cols]  # Get text and remove extra spaces
    data.append(cols)  # Add the row's data to the list

# Now, `data` is a list of lists, where each inner list represents a row of data.
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Hereâ€™s an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Cell 6: Create an empty list and loop through the scraped data to create a list of rows

# Create an empty list
data_list = []

# Loop through the scraped data to create a list of rows
for row in data:
    # Assuming each row has the following columns in order: id, terrestrial_date, sol, ls, month, min_temp, pressure
    if len(row) == 7:  # Only process rows with exactly 7 columns
        row_dict = {
            'id': row[0],
            'terrestrial_date': row[1],
            'sol': row[2],
            'ls': row[3],
            'month': row[4],
            'min_temp': row[5],
            'pressure': row[6]
        }
        data_list.append(row_dict)

```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
df = pd.DataFrame(data_list)

# Display the DataFrame
df.head()
```

```python
# Confirm DataFrame was created successfully
df
```

### Step 4: Prepare Data for Analysis

Examine the data types that are currently associated with each column. If necessary, cast (or convert) the data to the appropriate `datetime`, `int`, or `float` data types.

  > **Hint** You can use the Pandas `astype` and `to_datetime` methods to accomplish this task.


```python
# Examine data type of each column
print(df.dtypes)
```

```python
# Change data types for data analysis

# Examine data type of each column
print("Data types before conversion:")
print(df.dtypes)

# Convert columns to appropriate data types
df['id'] = df['id'].astype(int)
df['terrestrial_date'] = pd.to_datetime(df['terrestrial_date'])
df['sol'] = df['sol'].astype(int)
df['ls'] = df['ls'].astype(float)
df['month'] = df['month'].astype(int)
df['min_temp'] = df['min_temp'].astype(float)
df['pressure'] = df['pressure'].astype(float)

# Verify the data types after conversion
print("\nData types after conversion:")
print(df.dtypes)
```

```python
# Confirm type changes were successful by examining data types again
print(df.dtypes)
```

```python
# Find the unique months in the dataset
unique_months = df['month'].unique()
print(unique_months)
```

```python
# Group the DataFrame by 'month' and calculate the mean of 'min_temp' for each month
average_min_temp_by_month = df.groupby('month')['min_temp'].mean()

# Convert the result to a DataFrame for better readability
average_min_temp_by_month_df = average_min_temp_by_month.reset_index()

# Display the DataFrame
print(average_min_temp_by_month_df)
```

```python
# Cell 11: Plot the average minimum daily temperature by month

# Calculate the average minimum daily temperature by month (if not already done)
average_min_temp_by_month = df.groupby('month')['min_temp'].mean()

# Plot the results as a bar chart
plt.figure(figsize=(10, 6))
plt.bar(average_min_temp_by_month.index, average_min_temp_by_month.values, color='skyblue')
plt.xlabel('Month')
plt.ylabel('Average Minimum Temperature (Â°C)')
plt.title('Average Minimum Daily Temperature by Month on Mars')
plt.xticks(average_min_temp_by_month.index)  # Ensure all months are labeled
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Cell 12: Identify the coldest and hottest months in Curiosity's location

# Calculate the average minimum daily temperature by month (if not already done)
average_min_temp_by_month = df.groupby('month')['min_temp'].mean()

# Identify the coldest and hottest months
coldest_month = average_min_temp_by_month.idxmin()
hottest_month = average_min_temp_by_month.idxmax()

print(f"The coldest month is: Month {coldest_month} with an average minimum temperature of {average_min_temp_by_month[coldest_month]:.2f}Â°C")
print(f"The hottest month is: Month {hottest_month} with an average minimum temperature of {average_min_temp_by_month[hottest_month]:.2f}Â°C")

# Plot the results as a bar chart
plt.figure(figsize=(10, 6))
plt.bar(average_min_temp_by_month.index, average_min_temp_by_month.values, color='skyblue')
plt.xlabel('Month')
plt.ylabel('Average Minimum Temperature (Â°C)')
plt.title('Average Minimum Daily Temperature by Month on Mars')
plt.xticks(average_min_temp_by_month.index)  # Ensure all months are labeled
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Highlight the coldest and hottest months
plt.bar(coldest_month, average_min_temp_by_month[coldest_month], color='blue', label=f'Coldest Month: {coldest_month}')
plt.bar(hottest_month, average_min_temp_by_month[hottest_month], color='red', label=f'Hottest Month: {hottest_month}')
plt.legend()

plt.show()
```

```python
# 4. Average pressure by Martian month
# Group the DataFrame by 'month' and calculate the mean of 'pressure' for each month
average_pressure_by_month = df.groupby('month')['pressure'].mean()

# Convert the result to a DataFrame for better readability
average_pressure_by_month_df = average_pressure_by_month.reset_index()

# Display the DataFrame, optionally with a clear title
print("Average Pressure by Martian Month:")
print(average_pressure_by_month_df)
```

```python
# Calculate the average pressure by month (if not already done)
average_pressure_by_month = df.groupby('month')['pressure'].mean()

# Plot the results as a bar chart
plt.figure(figsize=(10, 6))
plt.bar(average_pressure_by_month.index, average_pressure_by_month.values, color='lightcoral') 
plt.xlabel('Month')
plt.ylabel('Average Pressure (Pa)')  
plt.title('Average Pressure by Month on Mars')
plt.xticks(average_pressure_by_month.index)  
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 5. How many terrestrial (earth) days are there in a Martian year?

#make plot
plt.figure(figsize=(12, 6))  
plt.plot(df['sol'], df['min_temp'], color='skyblue', linewidth=1)  

# Add labels and title
plt.xlabel('Number of terrestrial days', fontsize=12)
plt.ylabel('Minimum temperature', fontsize=12)
plt.title('Minimum Temperature vs Number of Terrestrial Days on Mars')

# Customize ticks and gridlines
plt.xticks(range(0, max(df['sol']) + 1, 250), fontsize=10) 
plt.yticks(range(-90, -60 + 1, 5), fontsize=10)  
plt.grid(axis='y', linestyle='--', alpha=0.7) 

# Show the plot
plt.show()
```

On average, the third month has the coldest minimum temperature on Mars, and the eighth month is the warmest. But it is always very cold there in human terms!



Atmospheric pressure is, on average, lowest in the sixth month and highest in the ninth.

The distance from peak to peak is roughly 1425-750, or 675 days. A year on Mars appears to be about 675 days from the plot. Internet search confirms that a Mars year is equivalent to 687 earth days.
